```python
import pandas as pd
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head(21)
```

```python
data.loc[data['whoAmI']== 'robot', 'number'] = '0'
data.loc[data['whoAmI']== 'human', 'number'] = '1'
data.columns
data[['number']]
```
